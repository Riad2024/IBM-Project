```python
import pandas as pd  # Import pandas for data manipulation
import yfinance as yf  # Import yfinance for stock data extraction
import matplotlib.pyplot as plt  # Import matplotlib for plotting

# Download GameStop stock data
gme = yf.Ticker('GME')
gme_data = gme.history(period="max")

# Reset the index to use 'Date' as a column
gme_data.reset_index(inplace=True)

# Ensure that Date is in datetime format
gme_data['Date'] = pd.to_datetime(gme_data['Date'])

# Function to make the plot
def make_graph(data, title):
    plt.figure(figsize=(8, 3))
    plt.plot(data['Date'], data['Close'], label='Stock Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.show()

# Plot GameStop stock data
make_graph(gme_data, 'Historical Share Price')


```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you have already loaded the Tesla stock data (tesla_data) with 'Date' and 'Close' columns
# Assuming you have already loaded the Tesla revenue data (tesla_revenue_df) with 'Date' and 'Revenue' columns

# Ensure 'Date' columns are in datetime format for both datasets
tesla_data['Date'] = pd.to_datetime(tesla_data['Date'])
tesla_revenue_df['Date'] = pd.to_datetime(tesla_revenue_df['Date'])

# Plot 1: Tesla Historical Share Price
def plot_tesla_price(data, title):
    plt.figure(figsize=(8,3))
    plt.plot(data['Date'], data['Close'], label='Tesla Stock Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.show()

# Plot 2: Tesla Historical Revenue
def plot_tesla_revenue(data, title):
    plt.figure(figsize=(8, 3))
    plt.plot(data['Date'], data['Revenue'], label='Tesla Revenue', color='green')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Revenue (in millions)')
    plt.legend()
    plt.show()

# Plot Tesla Stock Price (Date vs Price)
plot_tesla_price(tesla_data, 'Tesla Historical Share Price')

# Plot Tesla Revenue (Date vs Revenue)
plot_tesla_revenue(tesla_revenue_df, 'Tesla Historical Revenue')
```
